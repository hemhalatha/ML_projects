```python
from ultralytics import YOLO
model = YOLO('yolov8n.pt')
model.train(data='/content/drive/MyDrive/yolov8/data.yaml', epochs=50, imgsz=640,project='/content/drive/MyDrive/yolov8/training_runs',
    name='run1')
```
