```python
from ultralytics import YOLO

model = YOLO("yolov8m.pt")  # downloads automatically if not found
```

```python

# Step 4: Train the model
model.train(
    data="/content/alu-2/data.yaml",
    task="detect",# path to your dataset YAML
    epochs=120,                        # train longer for more learning
    imgsz=512,
    batch=8,
    name="yolov8m_high_recall",
    lr0=0.001,                         # learning rate
    lrf=0.01,                          # final LR factor
    momentum=0.937,
    weight_decay=0.0005,
    warmup_epochs=3,
    augment=True,                      # enable augmentations
    degrees=5,
    scale=0.1,
    shear=0.1,
    translate=0.1,
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    flipud=0.2,
    fliplr=0.5,
    # Focus on recall:
    # lower box confidence threshold during training evaluation
    conf=0.2,
    iou=0.45,
    patience=30,                       # early stop if no improvement
    verbose=True
)
```
